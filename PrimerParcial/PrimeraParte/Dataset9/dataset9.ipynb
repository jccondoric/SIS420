### 2. Exploracion del dataset y revision de tipos de datos

```python
# mostrar la informacion general del DataFrame
print('INFORMACIÓN DE TIPO DE DATOS')
data.info()
```

```python
print('=== TIPOS DE DATOS ===')
print(f'Numéricas:   {len(data.select_dtypes(include=["number"]).columns)}')
print(f'Categóricas: {len(data.select_dtypes(include=["object"]).columns)}')
```

```python
# 1. Calculamos nulos y tipos de datos
nulos = data.isnull().sum()
tipos = data.dtypes

# 2. Creamos el reporte combinando todo
reporte_nulos = pd.DataFrame({
    'Tipo': tipos,
    'Total': nulos,
    '%': ((nulos / len(data)) * 100).round(2)
})

# 3. Filtramos solo las que tienen nulos y ordenamos
reporte_final = reporte_nulos[reporte_nulos['Total'] > 0].sort_values(by='Total', ascending=False)

print(f"Total de registros: {len(data)}")
print("-" * 40)
print(reporte_final)
```

```python
# REPORTE DE NULOS 
nulos = data.isnull().sum()

reporte = pd.DataFrame({
    'Tipo': data.dtypes,
    'Total': nulos,
    '%': ((nulos / len(data)) * 100).round(2)
})

# Filtramos solo lo que falta y ordenamos
reporte_final = reporte[reporte['Total'] > 0].sort_values(by='Total', ascending=False)

print(f"Total registros: {len(data)}")
print("-" * 45)
print(reporte_final)

```

```python
# 1. Identificamos solo las columnas numéricas
cols_numericas = data.select_dtypes(include=['float64', 'int64']).columns

print("Rellenando nulos numéricos con la mediana...")

# 2. Aplicamos la mediana solo a las que tienen nulos
for col in cols_numericas:
    if data[col].isnull().sum() > 0:
        mediana = data[col].median()
        data[col] = data[col].fillna(mediana)
        print(f'  ✓ {col}: completada con {mediana}')

print("-" * 30)
print("Proceso terminado.")
```
