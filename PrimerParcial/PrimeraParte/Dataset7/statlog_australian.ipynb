# Dataset 07 — Statlog (Australian Credit Approval) — UCI

**Descripción:** Similar al dataset Credit Approval pero de origen australiano. Contiene solicitudes de tarjetas de crédito con 14 características (mixtas: continuas y discretas). Los atributos están anonimizados por confidencialidad. Es una versión ligeramente diferente que aparece en otros repositorios como "Credit Screening Database".

**Tarea:** Clasificación binaria — predecir si una solicitud de crédito es aprobada (1) o rechazada (0).

**Características:**
- m = 690 registros
- n = 14 características + 1 clase
- Variable objetivo: `A15` (0 = rechazado, 1 = aprobado)
- Sin valores nulos (ya preprocesado)
- Mix de variables continuas y discretas

**URL UCI:** https://archive.ics.uci.edu/dataset/143/statlog+australian+credit+approval

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
print('Librerías cargadas correctamente')
```

## 1. Carga del Dataset

```python
columnas = [f'A{i}' for i in range(1,15)] + ['clase']
ruta = '/content/drive/MyDrive/machine learning/datasets/australian.dat'
data = pd.read_csv(ruta, names=columnas, sep=' ')
print(f'Dimensiones: {data.shape}')
data.head()
```

## 2. Revisión de Tipos y Nulos

```python
print(data.dtypes)
print(f'\nTotal nulos: {data.isnull().sum().sum()}')
```

## 3. Preprocesamiento con Pandas

```python
df = data.copy()

# Codificar categóricas si existen
cols_cat = df.select_dtypes(include=['object']).columns
for col in cols_cat:
    df[col] = pd.Categorical(df[col]).codes

y = df['clase'].to_numpy()
X = df[[col for col in df.columns if col != 'clase']].to_numpy(dtype=np.float64)

print(f'X: {X.shape} | y: {y.shape}')
print(f'Aprobadas (1): {int((y==1).sum())} ({(y==1).mean()*100:.1f}%)')
print(f'Rechazadas (0):{int((y==0).sum())} ({(y==0).mean()*100:.1f}%)')
```

## 4. Estadísticas Descriptivas

```python
df.describe()
```

## 5. Visualizaciones

```python
fig, axes = plt.subplots(1, 3, figsize=(15,4))
axes[0].bar(['Rechazada (0)','Aprobada (1)'],[int((y==0).sum()),int((y==1).sum())],
            color=['tomato','steelblue'],edgecolor='black')
axes[0].set_title('Distribución de Clases')
axes[0].set_ylabel('Solicitudes')

axes[1].hist(df['A2'],bins=20,color='steelblue',edgecolor='black')
axes[1].set_title('Distribución de A2')

axes[2].hist(df['A3'],bins=20,color='green',edgecolor='black')
axes[2].set_title('Distribución de A3')
plt.tight_layout()
plt.show()
```

## 6. División 80/20

```python
np.random.seed(42)
idx = np.random.permutation(len(X))
corte = int(len(X)*0.8)
X_train,X_test = X[idx[:corte]],X[idx[corte:]]
y_train,y_test = y[idx[:corte]],y[idx[corte:]]
print(f'Entrenamiento: {X_train.shape[0]} | Prueba: {X_test.shape[0]}')
```
