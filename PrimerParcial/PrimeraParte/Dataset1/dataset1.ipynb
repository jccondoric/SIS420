De Cock artículo académico y al conjunto de datos asociado titulado "Ames, Iowa: Alternative to the Boston Housing Data as an End of Semester Regression Project", publicado por el profesor Dean De Cock en el Journal of Statistics Education

**Características principales:**
- **Número de observaciones:** 2930 casas
- **Número de variables:** 82 columnas (81 características + `SalePrice`)
- **Variable objetivo:** `SalePrice` (precio de venta de la casa en dólares)

```python
#cargando el dataset
data = pd.read_csv('AmesHousing.csv')
data 
```

### 2. Exploracion del dataset y revision de tipos de datos

```python
# mostrar la informacion general del DataFrame
print('INFORMACIÓN DE TIPO DE DATOS')
data.info()
```

```python
print('=== TIPOS DE DATOS ===')
print(f'Numéricas:   {len(data.select_dtypes(include=["number"]).columns)}')
print(f'Categóricas: {len(data.select_dtypes(include=["object"]).columns)}')
```

```python
nulos = data.isnull().sum()
# Filtramos, ordenamos y calculamos porcentaje al mismo tiempo
reporte_nulos = pd.DataFrame({
    'Total': nulos[nulos > 0],
    '%': (nulos[nulos > 0] / len(data)) * 100
}).sort_values(by='Total', ascending=False)

print(f"Total de registros: {len(data)}")
print(reporte_nulos)
```

```python
# Eliminamos columnas con demasiados nulos
cols_a_eliminar = ['Pool QC', 'Misc Feature', 'Alley', 'Fence']
data.drop(columns=cols_a_eliminar, inplace=True, errors='ignore')
```

eliminados ['Control de calidad de la piscina', 'Características diversas', 'Callejón', 'Valla']

```python
# 1. Imputar numéricos con mediana (Tu código está perfecto aquí)
cols_num = data.select_dtypes(include=[np.number]).columns
for col in cols_num:
    if data[col].isnull().sum() > 0:
        data[col] = data[col].fillna(data[col].median())

# 2. Imputar categóricos restantes con la moda y luego codificar
cols_cat = data.select_dtypes(include=['object']).columns
for col in cols_cat:
    # Solo usamos la moda si aún quedan nulos (los que no eran 'None')
    if data[col].isnull().sum() > 0:
        data[col] = data[col].fillna(data[col].mode()[0])
    
    # Convertimos a números (Label Encoding)
    data[col] = pd.Categorical(data[col]).codes

print(f'Nulos restantes: {data.isnull().sum().sum()}')
print(f'Columnas finales: {data.shape[1]}')
```

### 3. Separación de variables predictoras (X) y variable objetivo (y)

```python
# Separamos X e y
X = data.drop('SalePrice', axis=1).values      # todas las columnas menos SalePrice
y = data['SalePrice'].values.reshape(-1, 1)    # variable objetivo como columna

print(f'Dimensiones:')
print(f'X (features) : {X.shape}')
print(f'y (target)   : {y.shape}')
```

### 4. Normalización de características

```python
def featureNormalize(X):
    """Normaliza las características (media = 0, desviación estándar = 1)"""
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma          
    return X_norm, mu, sigma

# Aplicamos normalización
X_norm, mu, sigma = featureNormalize(X)

print("Normalización completada")
print(f"Media de las primeras 5 características: {mu[:5].round(4)}")
print(f"Desviación estándar de las primeras 5 características: {sigma[:5].round(4)}")
```

### 5. Agregar término de bias (intercepto)

```python
# Agregar columna de 1s al inicio (necesario para la hipótesis h(x) = X · θ)
m = X_norm.shape[0]
X_norm = np.concatenate([np.ones((m, 1)), X_norm], axis=1)

print(f"X_norm con bias agregado: {X_norm.shape}")
```

### 6. Guardar dataset preparado

```python
# 1. Unimos X_norm (con bias) + y (SalePrice)
prepared_data = np.concatenate([X_norm, y], axis=1)

# 2. Creamos nombres de columnas claros
column_names = ['Bias'] + list(data.drop('SalePrice', axis=1).columns)

# 3. Convertimos a DataFrame
prepared_df = pd.DataFrame(prepared_data, columns=column_names + ['SalePrice'])

# 4. Guardamos
prepared_df.to_csv('AmesHousing_prepared.csv', index=False)

print("Dataset preparado guardado correctamente como 'AmesHousing_prepared.csv'")
print(f"   Filas               : {prepared_df.shape[0]}")
print(f"   Columnas            : {prepared_df.shape[1]}")
print(f"   Incluye bias        : Sí (primera columna)")
print(f"   Features normalizadas: Sí")
```
